```python
def Continuos_univariate_Analysis(data, features, figsize=(12, 7), kde=True, bins = None):
    f1, (ax_box1, ax_hist1)= plt.subplots(nrows=2, sharex=True, gridspec_kw={"height_ratios":(0.25, 0.75)}, figsize=figsize)
    sns.color_palette("BuPu_r", as_cmap=True)
    sns.boxplot(data=data, x=features, ax= ax_box1, showmeans=True, color='violet')
    sns.histplot(data=data, x=features, ax=ax_hist1, kde=kde, bins= bins, color="orange") if bins else\
         sns.histplot(data=data, x=features, ax=ax_hist1, kde=kde, color="orange") 
    ax_hist1.axvline(data[features].mean(), color='green', linestyle="--")
    ax_hist1.axvline(data[features].median(), color='orange', linestyle="-")
```

```python
def discrete_univariate_analysis(data, features, perc=False, n=None):
    total = len(data[features])
    count  = data[features].nunique()
    if n is None:
        plt.figure(figsize=(count+1, 5))
    else:
        plt.figure(figsize = (n+1, 5))
    plt.xticks(rotation =90, fontsize=15)
    ax = sns.countplot(data = data, x = features, palette='flare', order = data[features].value_counts().index[:n].sort_values(ascending=False))
    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(100*p.get_height()/total)
        else:
            label = p.get_height()
        x = p.get_x() +p.get_width()/2
        y = p.get_height()
        ax.annotate(label, (x,y), ha ='center', va='center', size=12, xytext = (0,5), textcoords = "offset points")
    plt.show()
```

```python
discrete_univariate_analysis (dataset, "ongoing", perc=True)
```

```python
x=dataset.drop(['rating'], axis=1)
y = dataset["rating"]
x.info()
```

```python


x= pd.get_dummies (x, columns=x.select_dtypes (include=['object', 'category']).columns.tolist(), drop_first=True)
x.head()
```

```python

x.drop(columns= 'ongoing', inplace=True)
```

```python
x.info
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print("Number of samples for train:", X_train.shape[0])
print("Number of samples for test:", X_test.shape[0])
```

```python
imputer = SimpleImputer(strategy='mean')


X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)


lin_model = LinearRegression()
lin_model.fit(X_train_imputed, y_train)
```

```python
def model_performance(model, predictor, target):
    pred = model.predict(predictor)
    r2 = r2_score(target, pred)
    rmse = np.sqrt(mean_squared_error(target, pred))
    results = pd.DataFrame({
        "Root Mean Squared Error": [rmse],
        "R2 Score": [r2]
    }, index=["Model Performance"])
    return results
```

```python


print("Training Data Performance")
lin_model_train = model_performance(lin_model, X_train_imputed, y_train)
print(lin_model_train)
```

```python

imputer_test = SimpleImputer(strategy='mean')
X_test_imputed = imputer_test.fit_transform(X_test)
print("Test Data Performance")
lin_model_test=model_performance(lin_model, X_test_imputed, y_test)
lin_model_test
```
